```python
import pandas as pd
import numpy as np
```

```python
all_ages = pd.read_csv('all_ages.csv')
all_ages.head()
```

```python
recent_grads = pd.read_csv('recent_grads.csv')
recent_grads.head()
```

```python
def major_categorie_len():
    all_major = {}
    recent_major = {}
    all_categories = all_ages['Major_category'].unique()
    recent_categorie = recent_grads['Major_category'].unique()
    for category in all_categories:
        is_category = all_ages['Major_category'] == category
        all_major[category] = all_ages[is_category]['Total'].sum()
    for category in recent_categorie:
        is_category = recent_grads['Major_category'] == category
        recent_major[category] = recent_grads[is_category]['Total'].sum()
    return all_major , recent_major
```

```python
all_major , recent_major = major_categorie_len()
all_major
```

```python
recent_major_pivot = recent_grads.pivot_table(index="Major_category",values=['Total'],aggfunc=np.sum)
recent_major_pivot
```

```python
low_wage_jobs = recent_grads['Low_wage_jobs'].sum()
low_wage_jobs
```

```python
taux_low_wage = low_wage_jobs / recent_grads['Total'].sum()
taux_low_wage
```

```python
def taux_low_wage_category(data):
   
    categories= data['Major_category'].unique()
    taux_low = {}
    total = data['Total'].sum()
    for category in categories:
        is_category = data['Major_category'] == category
        taux_low[category] =  data[is_category]['Low_wage_jobs'].sum() / total * 100
    return taux_low    
    
```

```python
taux_low_category = taux_low_wage_category(recent_grads)
taux_low_category
```

```python
majors = recent_grads['Major'].unique()
recent_grads.head()
```

```python
majors_count = 0
for m in majors:
    majors_all = all_ages[all_ages['Major'] == m]
    majors_recent = recent_grads[recent_grads['Major']==m]
    taux_majors_all = majors_all['Unemployment_rate'].values
    taux_majors_recent = majors_recent['Unemployment_rate'].values
    if taux_majors_recent < taux_majors_all:
        majors_count +=1

print(majors_count)
        
        
```
