```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate.head(12)
```

```python
x = unrate['DATE'][:12]
y = unrate['UNRATE'][:12]
plt.plot(x,y)
plt.show()
```

```python
#rotation sur l'axe x
plt.xticks(rotation=90)
plt.plot(x,y)
plt.show()
```

```python
def crypto_cesar(mot):
    chaine = ""
    for c in mot:
        chaine += chr(ord(c)+3)
    return chaine

def decrypto_cesar(mot):
    chaine = ""
    for c in mot:
        chaine +=chr(ord(c)-3)
    return chaine
mot = decrypto_cesar('vdoxw')
```

```python
fig = plt.figure(figsize=(10,6))
for i in range(5):
    start = i*12
    end = (i+1)*12
    subset = unrate[start:end]
    plt.plot(subset['Month'],subset['UNRATE'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
unrate['Year'] = unrate['DATE'].dt.year
for i in range(5):
    start = i*12
    end = (i+1)*12
    subset = unrate[start:end]
    year = subset['Year'].unique()
    plt.plot(subset['Month'],subset['UNRATE'],label=year[0])
    plt.legend(loc='upper left')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
unrate['Year'] = unrate['DATE'].dt.year
for i in range(5):
    start = i*12
    end = (i+1)*12
    subset = unrate[start:end]
    year = subset['Year'].unique()
    plt.plot(subset['Month'],subset['UNRATE'],label=year[0])
    plt.legend(loc='upper left')
plt.xlabel('Moi')
plt.ylabel('Taux de chomage en %')
plt.title('EvoluTaux du taux chomage de 1948 a 1952 en %')
plt.show()
```
